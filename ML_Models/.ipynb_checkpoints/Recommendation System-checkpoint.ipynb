```python
#Dataframe manipulation library
import pandas as pd
#Math functions, we'll only need the sqrt function so let's import only that
from math import sqrt
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movies_df=pd.read_csv('datasets/recommendation_sys_data/movies.csv')
```

```python
ratings_df=pd.read_csv('datasets/recommendation_sys_data/ratings.csv')
```

```python
movies_df.head()
```

```python
import re
```

```python
movies_df['title']=movies_df['title'].str.replace(r"\(.*\)","")
```

```python
#Applying the strip function to get rid of any ending whitespace characters that may have appeared
movies_df['title'] = movies_df['title'].apply(lambda x: x.strip())
```

```python
movies_df.head()
```

```python
movies_df['genres']=movies_df['genres'].str.split('|')
```

```python
movies_df.head()
```

```python
movies_with_genres=movies_df.copy()
```

```python
for index,row in movies_with_genres.iterrows():
    for genre in row['genres']:
        movies_with_genres.at[index,genre]=1
        
movies_with_genres.fillna(0,inplace=True)
```

```python
movies_with_genres.head()
```

```python
ratings_df.head()
```

```python
ratings_df.drop('timestamp',axis=1,inplace=True)
```

```python
ratings_df.head()
```

```python
userInput = [
            {'title':'Breakfast Club, The', 'rating':5},
            {'title':'Toy Story', 'rating':3.5},
            {'title':'Jumanji', 'rating':2},
            {'title':"Pulp Fiction", 'rating':5},
            {'title':'Akira', 'rating':4.5}
         ] 
inputMovies = pd.DataFrame(userInput)
inputMovies
```

```python
inputMovies['title'].tolist()
```

```python
#Filtering out the movies by title
inputId = movies_df[movies_df['title'].isin(inputMovies['title'].tolist())]
inputId
```

```python
#Then merging it so we can get the movieId. It's implicitly merging it by title.
inputMovies = pd.merge(inputId, inputMovies)
```

```python
inputMovies
```

```python
#Dropping information we won't use from the input dataframe
inputMovies = inputMovies.drop('genres', 1)
```

```python
inputMovies
```

```python
usermovies=movies_with_genres[movies_with_genres['title'].isin(inputMovies['title'].tolist())]
usermovies
```

```python
usermovies.drop(['movieId','title','genres'],axis=1,inplace=True)
```

```python
usermovies.reset_index(drop=True,inplace=True)
```

```python
user_genre_table=usermovies.copy()
```

```python
user_genre_table.transpose()
```

```python
inputMovies['rating']
```

```python
user_profile=user_genre_table.transpose().dot(inputMovies['rating'])
user_profile
```

```python
user_profile.sum()
```

```python
movies_with_genres.head()
```

```python
movies_with_genres.set_index('movieId',inplace=True)
```

```python
movies_with_genres=movies_with_genres.drop('title',1).drop('genres',1)
```

```python
recommendation_movies=((movies_with_genres*user_profile).sum(axis=1))/user_profile.sum()
```

```python
recommendation_movies.sort_values(ascending=False,inplace=True)
```

```python
movies_df.loc[movies_df['movieId'].isin(recommendation_movies.head(5).keys()),'title']
```
